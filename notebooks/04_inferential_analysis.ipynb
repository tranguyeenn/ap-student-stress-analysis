## 04 — Inferential analysis

Welch two-sample t-test, Cohen's d, and 95% CI for the mean difference (STEM minus Humanities).


```python
from pathlib import Path

import numpy as np
import pandas as pd
from scipy import stats

DATA_PATH = Path('../data/cleaned/student_stress.csv')
df = pd.read_csv(DATA_PATH)

stem = df.loc[df['course_type'] == 'STEM', 'pss_score']
humanities = df.loc[df['course_type'] == 'Humanities', 'pss_score']
```

```python
t_stat, p_value = stats.ttest_ind(stem, humanities, equal_var=False)

n1, n2 = len(stem), len(humanities)
s1, s2 = stem.std(ddof=1), humanities.std(ddof=1)
pooled_sd = np.sqrt(((n1 - 1) * s1**2 + (n2 - 1) * s2**2) / (n1 + n2 - 2))
cohens_d = (stem.mean() - humanities.mean()) / pooled_sd

se = np.sqrt(s1**2 / n1 + s2**2 / n2)
df_welch = (s1**2 / n1 + s2**2 / n2) ** 2 / (
    (s1**2 / n1) ** 2 / (n1 - 1) + (s2**2 / n2) ** 2 / (n2 - 1)
)
t_crit = stats.t.ppf(0.975, df_welch)
mean_diff = stem.mean() - humanities.mean()
ci_low = mean_diff - t_crit * se
ci_high = mean_diff + t_crit * se

results = pd.Series({
    'stem_n': n1,
    'humanities_n': n2,
    'stem_mean': round(stem.mean(), 2),
    'humanities_mean': round(humanities.mean(), 2),
    'mean_diff_stem_minus_humanities': round(mean_diff, 2),
    'welch_t': round(t_stat, 3),
    'p_value': round(p_value, 4),
    'cohens_d': round(cohens_d, 3),
    'ci_95_low': round(ci_low, 2),
    'ci_95_high': round(ci_high, 2),
})
results
```

**Interpretation:** Report p-values and effect sizes together. A non-significant result in this sample does not rule out a meaningful difference in other settings.

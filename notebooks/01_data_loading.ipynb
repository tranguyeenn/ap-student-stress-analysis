## 01 — Data loading and validation


```python
from pathlib import Path

import pandas as pd

DATA_PATH = Path('../data/cleaned/student_stress.csv')
REQUIRED_COLUMNS = [
    'participant_id', 'course_type', 'pss_score',
    'source', 'stress_level', 'cohort',
]
```

```python
df_raw = pd.read_csv(DATA_PATH)
df_raw.head()
```

```python
df = df_raw.dropna(subset=['participant_id', 'course_type', 'pss_score']).copy()
df['participant_id'] = df['participant_id'].astype(int)
df['pss_score'] = df['pss_score'].astype(float)

missing_cols = set(REQUIRED_COLUMNS) - set(df.columns)
if missing_cols:
    raise ValueError(f'Missing columns: {missing_cols}')

df.info()
df.describe(include='all')
```

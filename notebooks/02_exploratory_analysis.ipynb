## 02 — Exploratory data analysis


```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')

DATA_PATH = Path('../data/cleaned/student_stress.csv')
df = pd.read_csv(DATA_PATH)
```

```python
df.isna().sum()
```

```python
df[['course_type', 'stress_level', 'cohort']].describe(include='object')
```

```python
pd.crosstab(df['course_type'], df['stress_level'], margins=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(data=df, x='pss_score', hue='course_type', kde=True, ax=ax)
ax.set_title('PSS score distribution in this sample')
ax.set_xlabel('PSS score')
plt.tight_layout()
plt.show()
```

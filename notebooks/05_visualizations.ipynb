## 05 — Visualizations

Figures are saved to `visuals/` for portfolio use.


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style='whitegrid')

DATA_PATH = Path('../data/cleaned/student_stress.csv')
VISUALS_DIR = Path('../visuals')
VISUALS_DIR.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(DATA_PATH)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(data=df, x='pss_score', hue='course_type', multiple='stack', ax=ax)
ax.set_title('Histogram of PSS scores in this sample')
ax.set_xlabel('PSS score')
fig.savefig(VISUALS_DIR / 'histogram_pss_by_course_type.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.boxplot(data=df, x='course_type', y='pss_score', ax=ax)
ax.set_title('PSS score by AP course category (this sample)')
ax.set_xlabel('Course type')
ax.set_ylabel('PSS score')
fig.savefig(VISUALS_DIR / 'boxplot_pss_by_course_type.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
means = df.groupby('course_type', as_index=False)['pss_score'].mean()

fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(data=means, x='course_type', y='pss_score', ax=ax, errorbar='sd')
ax.set_title('Mean PSS score by AP course category (this sample)')
ax.set_xlabel('Course type')
ax.set_ylabel('Mean PSS score')
fig.savefig(VISUALS_DIR / 'mean_pss_by_course_type.png', dpi=150, bbox_inches='tight')
plt.show()
```

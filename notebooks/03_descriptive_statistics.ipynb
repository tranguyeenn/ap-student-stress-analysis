## 03 — Descriptive statistics by AP course category


```python
from pathlib import Path

import pandas as pd

DATA_PATH = Path('../data/cleaned/student_stress.csv')
df = pd.read_csv(DATA_PATH)
```

```python
summary = (
    df.groupby('course_type')['pss_score']
    .agg(['count', 'mean', 'std', 'min', 'max'])
    .round(2)
    .rename(columns={'count': 'n'})
)
summary
```

```python
overall = df['pss_score'].agg(['count', 'mean', 'std', 'min', 'max']).round(2)
overall.name = 'all_participants'
overall
```
